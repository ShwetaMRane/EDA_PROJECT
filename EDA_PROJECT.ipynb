Multivariate Analysis

```python
# Box plot for AMT_CREDIT by TARGET (This is numerical multivariate)
plt.figure(figsize=(8, 6))
sns.boxplot(x='TARGET', y='AMT_CREDIT', data=df_app, palette='coolwarm')
plt.title("Credit Amount by Target Class")
plt.xlabel("Target")
plt.ylabel("AMT_CREDIT")
plt.show()

# Count plot of NAME_INCOME_TYPE grouped by TARGET
plt.figure(figsize=(10, 6))
sns.countplot(x='NAME_INCOME_TYPE', hue='TARGET', data=df_app, palette='viridis')
plt.title("Income Type Distribution by Target")
plt.xlabel("Income Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

# Grouped summary statistics by TARGET
target_summary = df_app.groupby('TARGET')[['AMT_ANNUITY', 'AMT_CREDIT']].mean()
print(target_summary)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='NAME_INCOME_TYPE', y='AMT_INCOME_TOTAL', hue='TARGET', data=df_app, palette='Set1', dodge=True)
plt.title("Income Distribution by Target and Income Type")
plt.xlabel("Income Type")
plt.ylabel("Total Income Amount")
plt.legend(title='Target')
plt.show()
```

```python
# Pair plot for numerical features, colored by TARGET
sns.pairplot(df_app[['AMT_INCOME_TOTAL', 'AMT_CREDIT', 'DAYS_BIRTH', 'TARGET']], hue='TARGET', palette='Set2')
plt.show()
```

```python
# 1. Multivariate Aggregation by TARGET and NAME_INCOME_TYPE
multivariate_summary = df_app.groupby(['TARGET', 'NAME_INCOME_TYPE'])[['AMT_ANNUITY', 'AMT_CREDIT']].mean()
print(multivariate_summary)

# Reset index for easier plotting
multivariate_summary = multivariate_summary.reset_index()

# 2. Clustered Bar Plot for AMT_ANNUITY and AMT_CREDIT by Income Type and Target
plt.figure(figsize=(14, 6))
sns.barplot(x='NAME_INCOME_TYPE', y='AMT_ANNUITY', hue='TARGET', data=multivariate_summary, palette='coolwarm')
plt.title("Average Annuity by Target and Income Type")
plt.xlabel("Income Type")
plt.ylabel("Average AMT_ANNUITY")
plt.xticks(rotation=45)
plt.legend(title="TARGET")
plt.show()

plt.figure(figsize=(14, 6))
sns.barplot(x='NAME_INCOME_TYPE', y='AMT_CREDIT', hue='TARGET', data=multivariate_summary, palette='viridis')
plt.title("Average Credit Amount by Target and Income Type")
plt.xlabel("Income Type")
plt.ylabel("Average AMT_CREDIT")
plt.xticks(rotation=45)
plt.legend(title="TARGET")
plt.show()

# 3. Violin Plot for AMT_INCOME_TOTAL Distribution by TARGET and NAME_INCOME_TYPE
plt.figure(figsize=(12, 6))
sns.violinplot(x='NAME_INCOME_TYPE', y='AMT_INCOME_TOTAL', hue='TARGET', split=True, data=df_app, palette='Set1')
plt.title("Income Distribution by Target and Income Type")
plt.xlabel("Income Type")
plt.ylabel("AMT_INCOME_TOTAL")
plt.xticks(rotation=45)
plt.legend(title="TARGET")
plt.show()
```

```python
correlation_matrix = df_app[['AMT_ANNUITY', 'AMT_CREDIT', 'AMT_INCOME_TOTAL', 'TARGET']].corr()

# 5. Heatmap for Correlation Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
# Sample DataFrame creation for demonstration
# df_app = pd.DataFrame({
#     'AMT_ANNUITY': [2000, 3000, 4000, 5000, 6000],
#     'AMT_CREDIT': [100000, 200000, 150000, 300000, 250000],
#     'AMT_INCOME_TOTAL': [40000, 50000, 60000, 70000, 80000],
#     'TARGET': [0, 1, 0, 1, 0]
# })

# Step 1: Aggregate the data
# Summing the numerical features grouped by TARGET
agg_data = df_app.groupby('TARGET')[['AMT_ANNUITY', 'AMT_CREDIT', 'AMT_INCOME_TOTAL']].sum().reset_index()

# Step 2: Prepare data for the pie chart
# Calculate the total amounts for each TARGET
sizes = agg_data[['AMT_ANNUITY', 'AMT_CREDIT', 'AMT_INCOME_TOTAL']].sum().values
labels = ['AMT_ANNUITY', 'AMT_CREDIT', 'AMT_INCOME_TOTAL']

# Step 3: Create the pie plot
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#66c2a5', '#fc8d62', '#8da0cb'])
plt.title('Pie Chart of Numerical Features')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the plot
plt.show()
```
